## I. Matplotlib 시작하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## II. Case Study with Arguments

```python
plt.plot([3, 6, 3, 6, 3])
plt.show()
```

```python
plt.figure(figsize = (9, 6))
plt.plot([2, 4, 2, 6, 3])
plt.show()
```

```python
# 이차함수 그리기 - 1
x = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
y = np.array([x**2 for x in x])
plt.plot(x, y)
plt.show()
```

```python
# 이치함수 그리기 - 2, label 추가
x = np.arange(-10, 10, 0.1)
y = np.array([x ** 2 for x in x])
###
plt.xlabel('x_value')
plt.ylabel('y_value')
###
plt.plot(x, y)
plt.show()
```

```python
# 이차함수 그리기 - 3, 범위 설정
x = np.arange(-10, 10, 0.1)
y = np.array([x ** 2 for x in x])

plt.xlabel('x_value')
plt.ylabel('y_value')
###
plt.axis([-5.5, 5.5, -3, 35]) # x 값의 최소, 최대, y값의 최소, 최대
###
plt.plot(x, y)
plt.show()
```

```python
# 이차함수 그리기 - 4. 눈금 그리기
x = np.arange(-10, 10, 0.1)
y = np.array([x ** 2 for x in x])

plt.xlabel('x_value')
plt.ylabel('y_value')

plt.axis([-5, 5, -3, 35]) # x 값의 최소, 최대, y값의 최소, 최대

###
plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(-3, 28, 3)])
###

plt.plot(x, y)
plt.show()
```

```python
# 이차함수 그리기 - 5. 타이틀 그리기, 범례
x = np.arange(-10, 10, 0.1)
y = np.array([x ** 2 for x in x])

plt.xlabel('x_value')
plt.ylabel('y_value')
plt.axis([-5, 5, -3, 35]) # x 값의 최소, 최대, y값의 최소, 최대
plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(-3, 28, 3)])
###
plt.title("y = x^2 graph")
plt.plot(x, y, label = "trend")
plt.legend()  # plot label을 지정 후에 결정함.
###
plt.show()
```

## III. Matplotlib Case Study

```python
x = np.arange(20)
y = np.random.randint(0, 20, 20)
plt.axis([0, 20, 0, 20])
plt.xticks([i for i in range(0, 20, 1)])
plt.yticks([0, 5, 10, 15, 20])
plt.xlabel('x_val')
plt.plot(x, y)
plt.show()
```

```python
# Scatter Graph
plt.scatter(x, y)
plt.show()
```

```python
# Box plot ( 사분위수 관련)
plt.boxplot(y)
plt.title('Box plot of y')
plt.show()
```

```python
#barplot
plt.bar(x, y)
plt.xticks([x for x in range(0, 20, 1)])
plt.show()
```

```python
plt.hist(y, bins = np.arange(0, 20, 2))
plt.xticks(np.arange(0, 20, 2))
plt.show()
```

```python
# Pie chart
# 비율 확인에 용이한 PieChart
z = [100, 400, 200, 300, 500]
plt.figure(figsize = (7, 7))
plt.pie(z, labels = ['a', 'b', 'c', 'd', 'NA'])
plt.legend(loc = 'upper right')
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

```python
import seaborn as sns
```

## Kdeplot

```python
# kernel density plot?
# sns.kdeplot()
x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)
plt.hist(y, bins = x)
plt.show()
```

```python
# smoothing한 데이터 보기?
# kdeplot을 써보자.
# 각 부분에 대한 밀도를 알 수 있다.
sns.kdeplot(y, shade =True)
plt.show()
```

## Countplot

```python
#Countplot, 빈도수의 시각화 groupby 도수 그리기.
vote_df = pd.DataFrame({"name" : ['Andy', 'Cathy', 'Tommy', 'Windy', 'Rossi'], "vote": [True, True, False, True, False]})
vote_df
```

```python
# matlibplot
vote_count = vote_df.groupby('vote').count()
vote_count
```

```python
plt.bar(x=[False, True], height = vote_count['name'])
plt.show()
```

```python
sns.countplot(x=vote_df['vote'])

plt.show()
```
