```python
# 조건별 계산
# 신규확진자 100 이상인 국가 추출하기.ㅠ
# loc를 활용하여 전체 데이터를 추출하는 방법.
# loc는 행의 조건, 열의 조건을 전부 추출한다.
# 즉 loc는 행과 열을 동시에 조건을 줄 때 편리하다.
covid.loc[covid['New cases']>=100,:'Country/Region']
```

### 행 기준 접근

```python
# Examples. - library
books_dict = {"Available":[True, True, False], "location":[102, 333, 254], "Genre":['Programming', 'Math', 'Physics']}
books = pd.DataFrame(books_dict, index =['Bug','Calculus','Mechnics'])
books
```

```python
# key를 준다.
books.loc["Bug",'Available']
```

```python
# index 로 준다. 0부터 시작하자.
books.iloc[1,1]
```

```python
books.iloc[1,0:2]
```

## Mission:
### 1. covid 데이터에서 100 case 대비 사망률(`Deaths / 100 Cases`)이 가장 높은 국가는?

```python
# sort를 통해 쉽게 알아낼 수 있다. 정답은 Yemen.
covid.sort_values(by = 'Deaths / 100 Cases', ascending = False)
```

### 2. covid 데이터에서 신규 확진자가 없는 나라 중 WHO Region이 'Europe'를 모두 출력하면?  
Hint : 한 줄에 동시에 두가지 조건을 Apply하는 경우 Warning이 발생할 수 있습니다.

```python
# 동시에 apply 하지말고 loc로 여러줄이 나오면, data frame 상태이므로 또  loc이 가능하다.
covid.loc[covid['WHO Region'] == 'Europe',:].loc[covid['New cases'] == 0,:]
```
